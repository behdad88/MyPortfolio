## Query by Username
Creation of queries using Tweepy API

Function is focused on completing the query then providing a CSV file of that query using pandas

```python
def username_tweets_to_csv(username,count):
    tweets = []
    counter =0
    print('Fetching {} tweets...'.format(username), '\n')
    try: 
    # Pulling individual tweets from query
        for tweet in api.user_timeline(id=username, count=count ):
            # Adding to list that contains all tweets
            counter+=1
            if counter==50:
                print('50 tweets downloaded...\n continue...\n')
            elif counter ==200:
                print('200 tweets downloaded...\n continue...\n')
            elif counter==500:
                print('More than 500 tweets downloaded...\n continue...\n')
            elif counter==1000:
                print('More than 500 tweets downloaded...\n continue...\n')
            
            tweets.append((tweet.created_at, tweet.id, tweet.text , tweet.retweet_count , tweet.favorite_count ,tweet.entities ,tweet.source, tweet.user.screen_name
                           , tweet.user.name,tweet.user.location, tweet.user.description, tweet.user.followers_count
                           , tweet.user.friends_count, tweet.user.created_at))
            
        # Creation of dataframe from tweets list
        tweetsdf = pd.DataFrame(tweets,columns=['Datetime','Tweet Id', 'Text','retweet_count' ,'favorite_count', 'entities','tweet_source'
                                                ,'user_screenname','user_name','location','user_description'
                                                ,'user_followers', 'user_friends', 'user_create_account'])

        # Converting dataframe to CSV
        tweetsdf.to_csv('{}-tweets.csv'.format(username)) 
        print('Saving {} tweets'.format(username , '\n'))

    except BaseException as e:
        print('failed on_status, user{}'.format(username),str(e), '\n')
        time.sleep(3)
```

## Query by Text Search
Function is focused on completing the query then providing a CSV file of that query using pandas

```python
tweets = []

def text_query_to_csv(text_query,count):
    try:
    # Pulling individual tweets from query
        for tweet in api.search(q=text_query, count=count):
            # Adding to list that contains all tweets
            tweets.append((tweet.created_at,tweet.id,tweet.text))
#             print(tweet)
            
        # Creation of dataframe from tweets list
        tweetsdf = pd.DataFrame(tweets,columns=['Datetime', 'Tweet Id', 'Text'])
            
        # Converting dataframe to CSV
        tweetsdf.to_csv('{}-tweets.csv'.format(text_query)) 

    except BaseException as e:
        print('failed on_status,',str(e))
        time.sleep(3)
```
