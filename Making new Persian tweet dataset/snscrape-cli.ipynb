```python
# Setting variables to be used in format string command below
tweet_count = 20000
shad_words = ['شادی', 'خوشحالم' ,'خوشحالی',  'سرور',  'شادمانی', 'مسرور' ,'مسرت',  'نشاط',  'جشن' , 'خوشحالیم' ,'خوشحالن'
              ,'عیش' , 'شادان', 'شاد', 'خندان','خندیدن','مسرورم', 'سرحال','خشنود', 'شادمان', 'رضایت', 'راضی' 
              ,'خوشم','خوش', 'خوب', 'سرخوش', 'خوشحالیم', 'خوشحالن', 'حال خوش', 'راضیم', 'خشنودم']
since_date = "2021-04-01"
until_date = "2021-07-01"

# Using OS library to call CLI commands in Python
for text_query in shad_words:
    print(text_query)
    os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> {}.json'.format(tweet_count, since_date, text_query, until_date , text_query))
```

```python
import json
import glob

temp = pd.DataFrame()

path_to_json = '/Joy' 

json_pattern = os.path.join(path_to_json,'*.json')
file_list = glob.glob(json_pattern)

dfs = [] # an empty list to store the data frames
for file in file_list:
    data = pd.read_json(file, lines=True) # read data frame from json file
    dfs.append(data) # append the data frame to the list

temp = pd.concat(dfs, ignore_index=True) # concatenate all the data frames in the list.
    
temp.head()
```
